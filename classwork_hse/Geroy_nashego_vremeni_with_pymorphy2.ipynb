```python
import os
import pymorphy2
import nltk
from nltk.tokenize import RegexpTokenizer
import re


tokenizer = RegexpTokenizer(r'\w+')
morph = pymorphy2.MorphAnalyzer()


lst = os.listdir('/Users/anyway/homework/classwork_hse/morph/texts')
print(lst)
```

```python
tokenizer = RegexpTokenizer(r'\w+')

parsed_words_pymorphy = []
for fl in lst:
    f = open('/Users/anyway/homework/classwork_hse/morph/texts'+ os.sep + fl, 'r', encoding='UTF-8')
    text = f.read().split()
    for word in text:
        word = tokenizer.tokenize(word)
        if type(word) == list:
            word = ''.join(word)
        if str(word) != '':
            parsed_word = morph.parse(word)[0]
            lemma = parsed_word.normal_form
            tag = parsed_word.tag.POS
            parsed_words_pymorphy.append([str(lemma), str(tag)])

for i in parsed_words_pymorphy:
    if i[1] == 'None':
        parsed_words_pymorphy.remove(i)
print(parsed_words_pymorphy[:50])
print(len(parsed_words_pymorphy))
```

```python
with open('/Users/anyway/homework/classwork_hse/morph/mystem_dataset.txt', 'r') as f:
    parsed_words_mystem = f.read().split('\n')
    print(parsed_words_mystem[:500])
```

```python
for i, word in enumerate(parsed_words_mystem):
    parsed_words_mystem[i] = word.split('\t')
print(parsed_words_mystem[:50])
print(len(parsed_words_mystem))
```

```python
lemmas_mystem = [str(i[0]) for i in parsed_words_mystem]
lemmas_pymorphy = [str(i[0]) for i in parsed_words_pymorphy]
for i in range(len(lemmas_pymorphy)):
    print(lemmas_pymorphy[i], lemmas_mystem[i])
```
